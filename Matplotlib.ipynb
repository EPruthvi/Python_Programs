```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.,5.,0.5)

plt.plot(t,t**2,'ro',label = '^2')
plt.plot(t,t**2.2,'b--',label = '^2.2')
plt.plot(t,t**2.4,'gs',label = '^2.4')
plt.legend()
plt.grid()
plt.title("Colortest")

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]

line = plt.plot(x1,y1,x2,y2)

plt.setp(line[0], color='blue', linewidth = 5, label = 'squares')
plt.setp(line[1], color='red', linewidth = 2, label = 'twos')
#plt.show()
plt.legend()
plt.grid()
plt.title("Lines sets")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(r):
    return np.pi*2*r

r1=np.arange(0.0,5.0,0.5)
r2 = np.arange(0.0,5.0,0.2)

plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(r1,f(r1),'r',label = 'Radius')
plt.title("Radius")
plt.legend()

plt.subplot(212)
plt.grid()
plt.plot(np.pi*r2 ** 2,'g')
plt.title("Area")

plt.show
```
